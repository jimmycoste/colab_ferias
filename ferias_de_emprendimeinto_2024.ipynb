## Definición de términos

**Politécnico**

En la República Dominicana, un politécnico es un centro educativo del nivel secundario que oferta la educación técnico profesional en más de un servicio de esta modalidad.

**Oferta**

Se refiere a los servicios ofertados en el Politécnico, digase, los *Titulos* o programas de *Bachilleratos técnicos* del Catálogo de Títulos de la Educación Técnico Profesional, debidamente implementados y con matrícula asignada.

**Bachillerato Técnico**

Diseño curricular para el segundo ciclo del nivel secundario en la modalidad técnico profesional, vinculado a perfiles profesionales y organizado por familias profesionales. Se recogen en el Catálogo de Títulos de la Educación Técnico Profesional. Tambien suele referirse a este como *Titulos* y constituye la oferta curricular de un politécnico, una vez está implementado y cuenta con matrícula asignada.

**Feria de emprendimiento.**

Una feria de emprendimiento es una actividad curricular que realiza un politécnico para mostrar mediante una propuesta realizable, diseñada y ejecutada por los estudiantes a partir de sus aprendizajes durante sus 3 años de educación y formación en la Educación Técnico Profesional, en sus respectivos titulos de Bachiller Técnico. Pueden participar estudiantes de los 3 grados, digase de 4to, 5to y 6to grado del segundo ciclo del nivel secundario.

Las ferias técnicas o de emprendimiento, están orientadas a estimular los aprendizajes obtenidos en el transcurso de su formación técnica, estos aprendizajes se evidencian mediante el reconocimiento de competencias relativas al desarrollo de la acción emprendedora y la simulación de la creación de una micro empresa en el aula.

## **Análisis**

**Preguntas:**

### ¿Cuántos politécnicos realizaron ferias de emprendimiento en el año escolar 2023-2024?
### ¿Cual es la proporción de politécnicos que realizaron ferias de emprendimiento en el año escolar 2023-2024?

```python
# Filtro politécnicos que realizarán ferias
ferias_realizaran_df = ferias_emprendimiento_df.dropna(subset=['FECHA_FERIA\n'])

# Número de politécnicos que realizarán ferias
num_politecnicos_ferias = ferias_realizaran_df['CENTRO'].nunique()

# Total de politécnicos
num_total_politecnicos = matricula_etp_df['POLITÉCNICO'].nunique()

# Proporción de politécnicos que realizarán ferias
proporcion_ferias = num_politecnicos_ferias / num_total_politecnicos

# Mostrar los resultados
print(f"Número de politécnicos que realizarán ferias: {num_politecnicos_ferias}")
print(f"Total de politécnicos: {num_total_politecnicos}")
print(f"Proporción de politécnicos que realizarán ferias: {proporcion_ferias:.2%}")
```

```python
# Etiquetas de los sectores
labels = ['Politécnicos con ferias', 'Politécnicos sin ferias']

# Tamaños de los sectores
sizes = [num_politecnicos_ferias, num_total_politecnicos - num_politecnicos_ferias]

# Colores de los sectores
colors = ["#00ffcc","#ff9900"]

# Crear el código pastel
plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Proporción de politécnicos que participan en ferias')
plt.show()
```

**Pregunta:**

### ¿En qué fechas se llevarán a cabo estas ferias?

```python
# Convertir el campo "FECHA_FERIA" a formato de fecha
ferias_emprendimiento_df['FECHA_FERIA\n'] = pd.to_datetime(ferias_emprendimiento_df['FECHA_FERIA\n'], format='%m/%d/%Y')

# Extraer el mes y el año de "FECHA_FERIA"
ferias_emprendimiento_df['MES'] = ferias_emprendimiento_df['FECHA_FERIA\n'].dt.month
ferias_emprendimiento_df['AÑO'] = ferias_emprendimiento_df['FECHA_FERIA\n'].dt.year

# Contar la cantidad de ferias por mes
ferias_por_mes = ferias_emprendimiento_df.groupby(['AÑO', 'MES']).size().reset_index(name='CANTIDAD_FERIAS')

# Mostrar los resultados
print("Cantidad de ferias por mes:")
print(ferias_por_mes)
```

```python
# Grafico de linea

sns.lineplot(data=ferias_por_mes, x='MES', y='CANTIDAD_FERIAS')
```

Con los resultados obtenidos de la cantidad de ferias por mes, podemos llegar a varias conclusiones:

**Estacionalidad de las ferias de emprendimiento:**

El mes con la mayor cantidad de ferias es mayo, seguido de abril. Esto puede indicar que hay una estacionalidad en la realización de las ferias de emprendimiento, con un pico de actividades en primavera, en donde se está en la antesala del fin de año escolar para los estudiantes Agosto - Junio.

**Necesidad de investigación adicional:**

La única feria registrada en enero y otra única realizada en septiembre podría ser un punto de interés para investigaciones adicionales. ¿Por qué solo hay una feria en enero? ¿Hay algún motivo específico para realizar la feria en ese mes? ¿Y porqué realizar una feria en septiembre, recien iniando el año escolar?

**Preguntas:**

### ¿Cuál es la matrícula y la oferta técnica de los Politécnicos que realizarán ferias?
### ¿Cuál es la proporción del impacto de las ferias de emprendimiento en los politécnicos y sus estudiantes de secundaria técnica a nivel nacional?

```python
# Obtener la matrícula de los politécnicos que realizarán ferias
matricula_ferias_df = matricula_etp_df[matricula_etp_df['SIGERD'].isin(ferias_realizaran_df['SIGERD'])]

# Calcular la matrícula total de estudiantes que participan en las ferias
matricula_total_ferias = matricula_ferias_df['MATRÍCULA'].sum()

# Calcular la matrícula total de estudiantes
matricula_total_etp = matricula_etp_df['MATRÍCULA'].sum()

# Calcular la proporción de estudiantes que participan en las ferias
proporcion_matricula_ferias = matricula_total_ferias / matricula_etp_df['MATRÍCULA'].sum()

# Mostrar los resultados
print(f"Matrícula total de estudiantes en el año escolar 2023-2024: {matricula_total_etp}")
print(f"Matrícula total de estudiantes que participan en las ferias: {matricula_total_ferias}")
print(f"Proporción de estudiantes que participan en las ferias: {proporcion_matricula_ferias:.2%}")
```

```python
# prompt: generar una visualizaión de pastel con seaborn para esta propoción

import seaborn as sns

# Create a pie chart
labels = ["Estudiantes participando", "Estudiantes no participando"]
sizes = [matricula_total_ferias, matricula_total_etp - matricula_total_ferias]
colors = ["#00ffcc","#ff9900"]

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%")
plt.title("Proporción de Estudiantes Participantes en Ferias")
plt.show()
```

Familias Profesionales y Títulos con incidencia en las ferias



```python
# Merge the dataframes on the common column 'SIGERD'
merged_df = pd.merge(ferias_emprendimiento_df, matricula_oferta_etp_df, on='SIGERD')

# Extract the Familia Profesional column
familia_profesional = merged_df['FAMILIA PROFESIONAL']

# Count the occurrences of each Familia Profesional
familia_profesional_count = familia_profesional.value_counts().sort_index()

# Get the matrícula for each Familia Profesional
matricula_familia_profesional = merged_df.groupby('FAMILIA PROFESIONAL')['MATRICULA'].sum()

# Create a DataFrame for the scatter plot
df_scatter = pd.DataFrame({
    'Familia Profesional': familia_profesional_count.index,
    'Conteo': familia_profesional_count.values,
    'Matricula': matricula_familia_profesional.values
})

print(df_scatter)

# Generate the scatter plot
plt.figure(figsize=(8, 8))
scatter_plot = sns.scatterplot(data=df_scatter, x='Conteo', y='Matricula', size='Matricula', hue='Familia Profesional', sizes=(70, 1000), alpha=0.7, palette='tab10')

# Añadir etiquetas y título
plt.title('Familias Profesionales con Mayor Presencia en Ferias', fontsize=16)
plt.xlabel('Número de Ferias', fontsize=14)
plt.ylabel('Total Matrícula', fontsize=14)
plt.legend(title='Familia Profesional', bbox_to_anchor=(1.05, 1), loc='upper left')

# Mostrar el gráfico
plt.show()
```
